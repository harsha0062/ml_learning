```python
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ml -working/train.csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
x=df.iloc[:,0:4]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
model = LogisticRegression()

scores = cross_val_score(model, x, y, scoring='accuracy', cv=20)

mean_value=float(np.mean(scores))
mean_value
```

**Applying Feature Construction**

```python
x['Family_size']=x['SibSp']+x['Parch']+1
```

```python
x.head()
```

```python
def myfunc(num):
  if num==1:
    # alone
    return 0
  elif num>1 and num<=4:
    # small family
    return 1
  else:
    # large family
    return 2
```

```python
x['Family_type']=x['Family_size'].apply(myfunc)
```

```python
x.head()
```

```python
x.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
x.head()
```

```python
from sklearn.linear_model import LogisticRegression
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
mean_value=float(np.mean(scores))
mean_value
```

**Feature Splitting**

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ml -working/train.csv')
```

```python
df.head()
```

```python
df['Name']
```

```python
df['Name'].str.split(',',expand=True)[1].str.split('.',expand=True)[0]
```

```python
df['Title']=df['Name'].str.split(',',expand=True)[1].str.split('.',expand=True)[0]
```

```python
df.groupby('Title')['Survived'].mean().sort_values(ascending=False)
```

```python
df['Is_Married']=0
df.loc[df['Title'] == 'Mrs', 'Is_Married'] = 1
df['Is_Married']
```
