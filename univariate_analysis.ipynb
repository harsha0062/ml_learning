```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("/content/train.csv")
```

```python
df.head()
```

```python
#1. categorical data
# count plot or bar graph
sns.countplot(x='Survived',data=df).plot(kind='bar')
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
# pie chart
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
# numerial data
# histogram
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

```python
# distplot
sns.distplot(df['Age'])
```

```python
# boxplot
sns.boxplot(x='Age',data=df)
```
